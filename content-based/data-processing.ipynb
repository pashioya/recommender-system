```python
import pandas as pd
import os

car_features = pd.read_csv('../datasets/car_features.csv')
```

```python
car_features.head()
```

```python
car_features.info()
```

```python
# Handling missing values
car_features['Market Category'] = car_features['Market Category'].fillna('UNKNOWN')

car_features['Number of Doors'] = car_features['Number of Doors'].fillna(2)

car_features['Engine HP'] = car_features['Engine HP'].fillna(car_features['Engine HP'].mean())

car_features['Engine Cylinders'] = car_features['Engine Cylinders'].fillna(car_features['Engine Cylinders'].mean())

car_features['Engine Fuel Type'] = car_features['Engine Fuel Type'].fillna('regular unleaded')
```

```python
# Fix Data Types
columns_to_convert = ['Make', 'Engine Fuel Type', 'Transmission Type', 'Driven_Wheels', 'Market Category', 'Vehicle Size', 'Vehicle Style','Model']
car_features[columns_to_convert] = car_features[columns_to_convert].astype('category')


```

```python
car_features.info()
```

```python
# Creating clean dataset csv
if os.path.exists('../datasets/clean_car_features.csv'):
    os.remove('../datasets/clean_car_features.csv')
    
car_features.to_csv('../datasets/clean_car_features.csv', index=False)
```
